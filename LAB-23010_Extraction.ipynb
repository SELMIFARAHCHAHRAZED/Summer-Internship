```python
import fitz
import re

# List of PDF file paths
pdf_paths = [
    "multiplication4_c_7kkh6i.pdf",
    "multiplication3_c_c56dllz.pdf",
    "multiplication5_c_88pody9i.pdf",
    "multiplication10_c_1s56zt.pdf"
    # Add more PDF file paths here
]

# Initialize a set to store multiplication operations
all_operations = set()
repeated_operations = set()

# Process each PDF
for pdf_path in pdf_paths:
    # Open the PDF using PyMuPDF
    pdf_document = fitz.open(pdf_path)

    # Process each page and extract text
    for page_number in range(pdf_document.page_count):
        page = pdf_document.load_page(page_number)
        page_text = page.get_text("text")

        # Use regular expression to extract multiplication operations
        multiplication_operations = re.findall(r'\d+\s*×\s*\d+', page_text)

        # Store operations in the set and find repeated ones
        for operation in multiplication_operations:
            if operation in all_operations:
                repeated_operations.add(operation)
            else:
                all_operations.add(operation)

    # Close the PDF
    pdf_document.close()

# Print repeated operations
if repeated_operations:
    print("Repeated operations:")
    for operation in repeated_operations:
        print(operation)
else:
    print("No repeated operations.")
```

```python

#MONGODB
import fitz
import re
from pymongo import MongoClient

# List of PDF file paths
pdf_paths = [
    "multiplication4_c_7kkh6i.pdf",
    "multiplication3_c_c56dllz.pdf",
    "multiplication5_c_88pody9i.pdf",
    "multiplication10_c_1s56zt.pdf"
    
]

# Initialize a list to store multiplication operations and their results
operations_and_results = []

# Process each PDF
for pdf_path in pdf_paths:
    # Open the PDF using PyMuPDF
    pdf_document = fitz.open(pdf_path)

    # Process each page and extract text
    for page_number in range(pdf_document.page_count):
        page = pdf_document.load_page(page_number)
        page_text = page.get_text("text")

        # Use regular expression to extract multiplication operations
        multiplication_operations = re.findall(r'\d+\s*×\s*\d+', page_text)

        # Store operations and their results
        for operation in multiplication_operations:
            operands = re.findall(r'\d+', operation)
            result = int(operands[0]) * int(operands[1])
            operations_and_results.append({"operation": operation, "result": result})

    # Close the PDF
    pdf_document.close()

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["Calculation"]
collection = db["Calculation"]

# Insert unique operations and results into the collection
for item in operations_and_results:
    collection.update_one(item, {"$set": item}, upsert=True)

# Print stored operations and results
for item in collection.find():
    print(f"{item['operation']} = {item['result']}")

# Print the total number of unique operations stored in the collection
print(f"Total number of unique operations stored in MongoDB: {collection.count_documents({})}")
```

```python
#mongodb
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page web
url = "https://www.atlas-monde.net/tous-les-pays/"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
content = response.content

# Utiliser Beautiful Soup pour parser le contenu HTML
soup = BeautifulSoup(content, "html.parser")

# Trouver le tableau contenant les informations
table = soup.find("table")

# Initialiser une liste pour stocker les données
geography_data_list = []

# Parcourir les lignes du tableau sauf la première (en-têtes)
for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if len(columns) >= 2:
        country = columns[0].text.strip()
        capital = columns[1].text.strip()
        geography_data_list.append({"Pays": country, "Capitale": capital})

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["INCLUSION"]
collection = db["Pays&Capitales"]

# Insérer les données dans la collection
collection.insert_many(geography_data_list)

# Afficher les données stockées
for data in collection.find():
    pays = data["Pays"]
    capitale = data["Capitale"]
    print(f"Pays : {pays}, Capitale : {capitale}")

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
#mongodb 
import PyPDF2
import pymongo

# Chemin vers le fichier PDF
pdf_path = "Comparatives_and_Superlatives_List.pdf"

# Fonction pour traiter les lignes du texte extrait
def process_lines(lines):
    data = []
    for line in lines:
        line = line.strip()
        if line:
            parts = line.split()
            if len(parts) == 3:
                data.append(parts)
    return data

# Ouvrir le PDF avec PyPDF2
with open(pdf_path, "rb") as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page_number in range(len(pdf_reader.pages)):
        text += pdf_reader.pages[page_number].extract_text()

# Diviser le texte en lignes et les traiter
lines = text.split('\n')
table_data = process_lines(lines)

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["TRANSFORMATION"]
collection = db["Comparatives&Superlatives"]

# Insérer les données dans la collection
for row in table_data[1:]:  # Ignorer le premier élément
    adjective, comparative, superlative = row
    data = {
        "Adjective": adjective,
        "Comparative": comparative,
        "Superlative": superlative
    }
    collection.insert_one(data)

# Afficher les données stockées
for data in collection.find():
    adjective = data["Adjective"]
    comparative = data["Comparative"]
    superlative = data["Superlative"]
    print(f"Adjective: {adjective}\nComparative: {comparative}\nSuperlative: {superlative}\n")

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page web
url = "https://engdic.org/20-irregular-adjectives-comparative-and-superlative-list/"

# Envoyer une requête HTTP et obtenir le contenu de la page
response = requests.get(url)
html_content = response.content

# Créer un objet BeautifulSoup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, "html.parser")

# Trouver tous les éléments <strong> qui contiennent les informations
strong_elements = soup.find_all("strong")

# Initialiser une liste pour stocker les données
data = []

# Parcourir les éléments <strong>
for strong in strong_elements:
    text = strong.get_text().strip()
    # Ignorer certaines lignes
    if text.lower() != "examples table of irregular adjectives comparative and superlative" and text.lower() != "positive" and text.lower() != "comparative" and text.lower() != "superlative" and text.lower() != "examples":
        data.append(text)

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["TRANSFORMATION"]
collection = db["IrregularAdjectivesC/S"]

# Insérer les données dans la collection
for item in data:
    data_entry = {"Adjective": item}
    collection.insert_one(data_entry)

# Afficher les données stockées
for data_entry in collection.find():
    adjective = data_entry["Adjective"]
    print(f"Adjective: {adjective}")

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
#Mongodb 
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page contenant le tableau
url = "https://englishan.com/contractions/"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
html_content = response.content

# Utiliser BeautifulSoup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, "html.parser")

# Trouver le tableau dans la page
table = soup.find("table")

# Parcourir les lignes du tableau
contractions_data = []

for row in table.find_all("tr")[1:]:  # Ignorer la première ligne d'en-tête
    cells = row.find_all("td")
    if len(cells) >= 2:  # S'assurer qu'il y a au moins 2 cellules dans la ligne
        contraction = cells[0].text.strip()
        expansion = cells[1].text.strip()
        contractions_data.append({"Contraction": contraction, "Expansion": expansion})

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["TRANSFORMATION"]
collection = db["Contractions"]

# Insérer les données dans la collection
collection.insert_many(contractions_data)

# Afficher les données stockées
for data_entry in collection.find():
    contraction = data_entry["Contraction"]
    expansion = data_entry["Expansion"]
    print(f"Contraction: {contraction}\nExpansion: {expansion}\n")

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
#Mongodb 
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient

# URL du site
url = "http://flemotion.com/emotions-francais-fle.html"

# Faire une requête GET au site
response = requests.get(url)

# Analyser le contenu HTML
soup = BeautifulSoup(response.content, "html.parser")

# Trouver le tableau contenant les informations
table = soup.find("tbody")

# Liste pour stocker les données de vocabulaire
vocabulaire = []

# Parcourir les lignes du tableau
for row in table.find_all("tr"):
    cells = row.find_all("td")
    if len(cells) >= 3:
        emotion = cells[0].text.strip()
        masculin = cells[1].text.strip()
        feminin = cells[2].text.strip()
        vocabulaire.append({"emotion": emotion, "masculin": masculin, "feminin": feminin})

# Se connecter à la base de données MongoDB
client = MongoClient("mongodb://localhost:27017/")  # Modifier l'URL de connexion si nécessaire
db = client["ASSOCIATION"]  # Nom de la base de données
vocabulaire_collection = db["ChampsLexEmotions"]  # Nom de la collection

# Insérer les données de vocabulaire dans la collection
vocabulaire_collection.insert_many(vocabulaire)

# Afficher les données de vocabulaire stockées dans la collection
for item in vocabulaire_collection.find():
    print(item)

# Fermer la connexion à la base de données
client.close()
```

```python
import requests
from bs4 import BeautifulSoup

# URL du site
url = "https://sherpas.com/blog/liste-des-homonymes/"

# Faire une requête GET au site
response = requests.get(url)

# Analyser le contenu HTML
soup = BeautifulSoup(response.content, "html.parser")

# Trouver les éléments <h4> avec la classe spécifiée
h4_elements = soup.find_all("h4", class_="wp-block-heading")

# Parcourir les éléments <h4> et afficher leur contenu
for h4 in h4_elements:
    print(h4.get_text())
```

```python
#mongodb 
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient

# URL du site
url = "https://sherpas.com/blog/liste-des-homonymes/"

# Faire une requête GET au site
response = requests.get(url)

# Analyser le contenu HTML
soup = BeautifulSoup(response.content, "html.parser")

# Trouver les éléments <h4> avec la classe spécifiée
h4_elements = soup.find_all("h4", class_="wp-block-heading")

# Liste pour stocker les données de vocabulaire
vocabulaire = []

# Parcourir les éléments <h4> et ajouter leur contenu à la liste
for h4 in h4_elements:
    vocabulaire.append(h4.get_text())

# Se connecter à la base de données MongoDB
client = MongoClient("mongodb://localhost:27017/")  # Modifier l'URL de connexion si nécessaire
db = client["ASSOCIATION"]  # Nom de la base de données
vocabulaire_collection = db["Homophones"]  # Nom de la collection

# Convertir les éléments de vocabulaire en documents de type dictionnaire
vocabulaire_documents = [{"mot": item} for item in vocabulaire]

# Insérer les documents de vocabulaire dans la collection
vocabulaire_collection.insert_many(vocabulaire_documents)

# Afficher les données de vocabulaire stockées dans la collection
for item in vocabulaire_collection.find():
    print(item)

# Fermer la connexion à la base de données
client.close()
```

```python
#mongodb +probleme affichage 
import requests
from bs4 import BeautifulSoup
import pymongo

# URL du site web à partir duquel vous souhaitez extraire les informations
url = "https://www.laculturegenerale.com/homophones-definition-signification-liste/"

# Effectuer une requête GET pour obtenir le contenu HTML de la page
response = requests.get(url)
html_content = response.content

# Utiliser Beautiful Soup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Initialiser une liste pour stocker les associations
association_list = []

# Trouver tous les éléments <li> dans la page
li_elements = soup.find_all('li')

# Parcourir les éléments <li> et stocker les associations dans la liste
for li in li_elements:
    text = li.get_text(strip=True)
    if text and "Pour résumer" not in text:
        association_list.append(text)

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["ASSOCIATION"]
collection = db["associations"]

# Insérer les associations dans la collection
for association in association_list:
    collection.insert_one({"association": association})

# Afficher les associations stockées
for association in collection.find():
    print(association["association"])
```

```python
from bs4 import BeautifulSoup
import requests

# URL du site web à partir duquel vous souhaitez extraire les informations
url = "https://la-liste.fr/langues/francais/les-mots-palindromes"

# Effectuer une requête GET pour obtenir le contenu HTML de la page
response = requests.get(url)
html_content = response.content

# Utiliser Beautiful Soup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Trouver tous les éléments <tr> dans la page
tr_elements = soup.find_all('tr')

# Initialiser une liste pour stocker les données
data_list = []

# Parcourir les éléments <tr> et extraire les données des colonnes <td>
for tr in tr_elements:
    td_elements = tr.find_all('td')
    if len(td_elements) == 2:
        col1 = td_elements[0].text.strip()
        col2 = td_elements[1].text.strip()
        data_list.append((col1, col2))

# Afficher les données extraites
for data in data_list:
    print(data)
```

```python
# MONGODB 
from bs4 import BeautifulSoup
import requests
import pymongo

# URL du site web à partir duquel vous souhaitez extraire les informations
url = "https://la-liste.fr/langues/francais/les-mots-palindromes"

# Effectuer une requête GET pour obtenir le contenu HTML de la page
response = requests.get(url)
html_content = response.content

# Utiliser Beautiful Soup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Trouver tous les éléments <tr> dans la page
tr_elements = soup.find_all('tr')

# Initialiser une liste pour stocker les données
data_list = []

# Parcourir les éléments <tr> et extraire les données des colonnes <td>
for tr in tr_elements:
    td_elements = tr.find_all('td')
    if len(td_elements) == 2:
        col1 = td_elements[0].text.strip()
        col2 = td_elements[1].text.strip()
        data_list.append({"col1": col1, "col2": col2})

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["ASSOCIATION"]
collection = db["Palindrome"]

# Insérer les données dans la collection
collection.insert_many(data_list)

# Afficher les données extraites et stockées
for data in collection.find():
    print(data)
```
