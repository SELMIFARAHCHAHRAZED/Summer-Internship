```python
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page web
url = "https://www.atlas-monde.net/tous-les-pays/"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
content = response.content

# Utiliser Beautiful Soup pour parser le contenu HTML
soup = BeautifulSoup(content, "html.parser")

# Trouver le tableau contenant les informations
table = soup.find("table")

# Initialiser une liste pour stocker les données
geography_data_list = []

# Parcourir les lignes du tableau sauf la première (en-têtes)
for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if len(columns) >= 2:
        country = columns[0].text.strip()
        capital = columns[1].text.strip()
        geography_data_list.append({"Keywords": "5eme année primaire geographie pays&capital en français",
                                   "Card1": f"Pays : {country}",
                                   "Card2": f"Capitale : {capital}"})

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["INCLUSION"]
collection = db["Pays&CapitalesV2"]

# Insérer les données dans la collection
collection.insert_many(geography_data_list)

# Afficher les données stockées
for data in collection.find():
    print(data)

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page web
url = "https://www.coursfrancaisfacile.com/2021/01/contraire-des-mots-francais-facile.html"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
content = response.content

# Utiliser Beautiful Soup pour parser le contenu HTML
soup = BeautifulSoup(content, "html.parser")

# Trouver tous les éléments <li> qui contiennent les antonymes
antonym_elements = soup.find_all('li')

# Initialiser une liste pour stocker les données
antonyms_list = []

# Parcourir chaque élément et stocker les mots et leurs antonymes
for element in antonym_elements:
    antonyms = element.text
    antonym_pair = antonyms.split('/')
    if len(antonym_pair) == 2:
        word = antonym_pair[0].strip()
        antonym = antonym_pair[1].strip()
        antonyms_list.append({"_Keywords": "Divers niveaux français vocabulaire antonyme&contraires",
                              "Card1": word,
                              "Card2": antonym})

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["TRANSFORMATION"]
collection = db["Antonymes"]

# Insérer les données dans la collection
collection.insert_many(antonyms_list)

# Afficher les données stockées
for data in collection.find():
    print(data)

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
#COMPERATIVE&SUPERLATIVE
```

```python
import PyPDF2
import pymongo

# Chemin vers le fichier PDF
pdf_path = "Comparatives_and_Superlatives_List.pdf"

# Fonction pour traiter les lignes du texte extrait
def process_lines(lines):
    data = []
    for line in lines:
        line = line.strip()
        if line:
            parts = line.split()
            if len(parts) == 3:
                data.append(parts)
    return data

# Ouvrir le PDF avec PyPDF2
with open(pdf_path, "rb") as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page_number in range(len(pdf_reader.pages)):
        text += pdf_reader.pages[page_number].extract_text()

# Diviser le texte en lignes et les traiter
lines = text.split('\n')
table_data = process_lines(lines)

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["TRANSFORMATION"]
collection = db["Comparatives&Superlatives"]

# Insérer les données dans la collection
for row in table_data[1:]:  # Ignorer le premier élément
    adjective, comparative, superlative = row
    data = {
        "_Keywords": "6eme primaire Anglais Comparatives&Superlatives",
        "Card1": {"Adjective": adjective},
        "Card2": {"Comparative": comparative},
        "Card3": {"Superlative": superlative}
    }
    collection.insert_one(data)

# Afficher les données stockées
for data in collection.find():
    _id = data["_id"]
    keywords = data["_Keywords"]
    card1 = data["Card1"]
    card2 = data["Card2"]
    card3 = data["Card3"]
    
    print(f"_id: {_id}")
    print(f"_Keywords: {keywords}")
    print(f"Card1: {card1}")
    print(f"Card2: {card2}")
    print(f"Card3: {card3}")
    print()

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
import pymongo

# Lien vers la page contenant le tableau
url = "https://englishan.com/contractions/"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
html_content = response.content

# Utiliser BeautifulSoup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, "html.parser")

# Trouver le tableau dans la page
table = soup.find("table")

# Parcourir les lignes du tableau
contractions_data = []

for row in table.find_all("tr")[1:]:  # Ignorer la première ligne d'en-tête
    cells = row.find_all("td")
    if len(cells) >= 2:  # S'assurer qu'il y a au moins 2 cellules dans la ligne
        contraction = cells[0].text.strip()
        expansion = cells[1].text.strip()
        contractions_data.append({
            "_Keywords": "6eme primaire Anglais Contraction",
            "Card1": {"Contraction": contraction},
            "Card2": {"Expansion": expansion}
        })

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["TRANSFORMATION"]
collection = db["Contractions"]

# Insérer les données dans la collection
collection.insert_many(contractions_data)

# Afficher les données stockées
for data_entry in collection.find():
    _id = data_entry["_id"]
    keywords = data_entry["_Keywords"]
    contraction = data_entry["Card1"]["Contraction"]
    expansion = data_entry["Card2"]["Expansion"]
    print(f"_id: {_id}")
    print(f"_Keywords: {keywords}")
    print(f"Card1:Contraction {contraction}")
    print(f"Card2:Expansion {expansion}")
    print()

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
import pymongo

# URL du site
url = "https://www.facile-anglais.com/index.php/cours-et-lecons-apprendre-anglais/fiches-utiles-memos-anglais/120-liste-de-plus-de-500-contraires-en-anglais"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
html_content = response.content

# Utiliser Beautiful Soup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Trouver le tableau contenant les informations
table = soup.find('table', class_='category')

# Initialiser des listes pour stocker les données
anglais_list = []
francais_list = []

# Parcourir les lignes du tableau
for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) >= 4:
        anglais = cols[1].get_text(strip=True)
        francais = cols[3].get_text(strip=True)
        anglais_list.append(anglais)
        francais_list.append(francais)

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["TRANSFORMATION"]
collection = db["TRADUCTION-FR/ANG"]

# Créer une liste de dictionnaires pour les données
data_list = [{"Anglais": anglais, "Français": francais} for anglais, francais in zip(anglais_list, francais_list)]

# Insérer les données dans la collection
collection.insert_many(data_list)

# Afficher les données stockées
for data in collection.find():
    _id = data["_id"]
    keywords = "Divers niveaux langue français et anglais vocabulaire traduction"
    card1 = f'Anglais\n"{data["Anglais"]}"'
    card2 = f'Français\n"{data["Français"]}"'
    
    # Mettre à jour les données dans la collection en enlevant les clés Anglais et Français
    collection.update_one({"_id": _id}, {"$set": {"_Keywords": keywords, "Card1": card1, "Card2": card2},
                                         "$unset": {"Anglais": "", "Français": ""}})
    
    print(f'_id\n{_id}')
    print(f'_Keywords\n"{keywords}"')
    print(f'Card1:{card1}')
    print(f'Card2:{card2}')
    print()

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
import pymongo

# URL du site
url = "https://www.facile-anglais.com/index.php/cours-et-lecons-apprendre-anglais/fiches-utiles-memos-anglais/120-liste-de-plus-de-500-contraires-en-anglais"

# Faire une requête GET pour obtenir le contenu de la page
response = requests.get(url)
html_content = response.content

# Utiliser Beautiful Soup pour analyser le contenu HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Trouver le tableau contenant les informations
table = soup.find('table', class_='category')

# Initialiser des listes pour stocker les données
data_list = []

# Parcourir les lignes du tableau
for row in table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) >= 4:
        anglais = cols[1].get_text(strip=True)
        antonyme_anglais = cols[0].get_text(strip=True)
        francais = cols[3].get_text(strip=True)
        antonyme_francais = cols[2].get_text(strip=True)
        data_list.append({
            "_Keywords": "Divers niveaux langue français et anglais vocabulaire traduction/Antonymes",
            "Card1": f"Anglais\"{anglais}\"",
            "Card2": f"Antonyme Anglais\"{antonyme_anglais}\"",
            "Card3": f"Français\"{francais}\"",
            "Card4": f"Antonyme Français\"{antonyme_francais}\""
        })

# Connexion à la base de données MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["TRANSFORMATION"]
collection = db["Traduction&ContrairesV2"]

# Insérer les données dans la collection
collection.insert_many(data_list)

# Afficher les données stockées
for data in collection.find():
    _id = data["_id"]
    keywords = data["_Keywords"]
    card1 = data["Card1"]
    card2 = data["Card2"]
    card3 = data["Card3"]
    card4 = data["Card4"]
    print(f"_id: {_id}")
    print(f"_Keywords: {keywords}")
    print(f"Card1:{card1}")
    print(f"Card2:{card2}")
    print(f"Card3:{card3}")
    print(f"Card4:{card4}")
    print()

# Afficher le nombre total d'entrées stockées
total_entries = collection.count_documents({})
print(f"Nombre total d'entrées stockées : {total_entries}")
```

```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient

# URL du site
url = "https://www.podcastfrancaisfacile.com/vocabulaire/corps-humain.html"

# Effectuer la requête HTTP GET
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Utiliser Beautiful Soup pour analyser le contenu de la page
    soup = BeautifulSoup(response.content, "html.parser")

    # Trouver tous les éléments <tr> contenant les lignes du tableau
    rows = soup.find_all("tr")

    # Liste pour stocker les données de vocabulaire
    vocabulaire = []

    # Parcourir les lignes du tableau (à partir de la deuxième ligne pour éviter l'en-tête)
    for row in rows[1:]:
        # Trouver les éléments <td> correspondant aux première et troisième colonnes
        cells = row.find_all("td")
        if len(cells) >= 3:  # S'assurer qu'il y a au moins 3 cellules dans la ligne
            partie_du_corps = cells[0].get_text().strip()
            traduction_anglais = cells[2].get_text().strip()
            data = {
                "_Keywords": "6eme année primaire langue français et anglais vocabulaire Partie du corps traduction",
                "Card1": partie_du_corps,
                "Card2": traduction_anglais
            }
            vocabulaire.append(data)

    # Se connecter à la base de données MongoDB
    client = MongoClient("mongodb://localhost:27017/")  # Modifier l'URL de connexion si nécessaire
    db = client["ASSOCIATION"]  # Nom de la base de données
    vocabulaire_collection = db["ChampsLexPartiesDuCorps+TraductionAngV2"]  # Nom de la collection

    # Insérer les données de vocabulaire dans la collection
    vocabulaire_collection.insert_many(vocabulaire)

    # Afficher les données de vocabulaire stockées dans la collection
    for data in vocabulaire_collection.find():
        print(f"_id: {data['_id']}")
        print(f"_Keywords: {data['_Keywords']}")
        print(f"Card1:{data['Card1']}")
        print(f"Card2:{data['Card2']}")
        print()

    # Fermer la connexion à la base de données
    client.close()
else:
    print("La requête a échoué avec le code de statut:", response.status_code)
```

```python
#HOMOPHONES
```

```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient

# URL du site
url = "https://sherpas.com/blog/liste-des-homonymes/"

# Faire une requête GET au site
response = requests.get(url)

# Analyser le contenu HTML
soup = BeautifulSoup(response.content, "html.parser")

# Trouver les éléments <h4> avec la classe spécifiée
h4_elements = soup.find_all("h4", class_="wp-block-heading")

# Liste pour stocker les données de vocabulaire
vocabulaire = []

# Parcourir les éléments <h4> et ajouter leur contenu à la liste
for h4 in h4_elements:
    vocabulaire.append(h4.get_text())

# Se connecter à la base de données MongoDB
client = MongoClient("mongodb://localhost:27017/")  # Modifier l'URL de connexion si nécessaire
db = client["ASSOCIATION"]  # Nom de la base de données
vocabulaire_collection = db["HomophonesV2"]  # Nom de la collection

# Initialiser une liste pour stocker les données formatées
formatted_data = []

# Parcourir les éléments du vocabulaire
for item in vocabulaire:
    mots = item.split(" / ")
    formatted_data.append({
        "_Keywords": "6eme année primaire langue français vocabulaire Homophones : même prononciation, écriture et sens différents",
        **{f"Card{i}": mot for i, mot in enumerate(mots, start=1)}
    })

# Insérer les données formatées dans la collection
vocabulaire_collection.insert_many(formatted_data)

# Afficher les données de vocabulaire stockées dans la collection
for data in vocabulaire_collection.find():
    print(f"_id: {data['_id']}")
    print(f"_Keywords: {data['_Keywords']}")
    card_number = 1
    while f"Card{card_number}" in data:
        print(f"Card{card_number}: {data[f'Card{card_number}']}")
        card_number += 1
    print()

# Fermer la connexion à la base de données
client.close()
```
